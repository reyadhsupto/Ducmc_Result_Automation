```python
#importing librarires
import pandas as pd

from selenium.webdriver.common.action_chains import ActionChains
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait

options = Options()
options.add_argument("--headless")
options.add_experimental_option("useAutomationExtension", False)
options.add_argument("--ignore-certificate-errors")
options.add_argument('disable-infobars')
options.add_argument('--disable-gpu')
options.add_argument("--start-maximized")
options.add_argument("--disable-notifications")

def wait_function_until_stale():
    WebDriverWait(driver, 100).until(lambda driver: driver.execute_script('return document.readyState') == 'complete')

main_page = f"https://ducmc.com/result.php"
search_key='bangladesh'

driver = webdriver.Chrome(options=options)
driver.get(main_page)
wait_function_until_stale()
webdriver.ActionChains(driver).send_keys(Keys.ESCAPE).perform()

#global wait 10 seconds
driver.implicitly_wait(10)

student_gpa = []
student_cgpa = []
final_reg = []
def gpa_extractor():
    gpa_divs = driver.find_elements(By.XPATH, "//td/div[contains(text(),'Promoted')]/small")
    student_gpa.append(gpa_divs[0].text.strip())
    print(gpa_divs[0].text.strip())
    student_cgpa.append(gpa_divs[1].text.strip())
    print(gpa_divs[1].text.strip())

def result_extractor(reg_no):
    #clicking search button and searching bangladeshi articles
    actions = ActionChains(driver)
    css_elem = f"#reg_no"
    reg_but = WebDriverWait(driver,25).until(EC.presence_of_element_located((By.CSS_SELECTOR, css_elem)))
    actions.move_to_element(reg_but).click().send_keys(reg_no).perform()


    xpath_elem = f"//select[@id='pro_id']"
    program_button = WebDriverWait(driver,25).until(EC.presence_of_element_located((By.XPATH, xpath_elem))).click()
    select_program = driver.find_element(By.XPATH , '//select[@id="pro_id"]/option[@value="14"]').click()

    if len(str(reg_no))==3:
        xpath_elem = f"//select[@id='sess_id']"
        session_button = WebDriverWait(driver,25).until(EC.presence_of_element_located((By.XPATH, xpath_elem))).click()
        select_session = driver.find_element(By.XPATH , '//select[@id="sess_id"]/option[@value="19"]').click()
        wait_function_until_stale()
    elif len(str(reg_no))==4:
        xpath_elem = f"//select[@id='sess_id']"
        session_button = WebDriverWait(driver,25).until(EC.presence_of_element_located((By.XPATH, xpath_elem))).click()
        select_session = driver.find_element(By.XPATH , '//select[@id="sess_id"]/option[@value="18"]').click()
        wait_function_until_stale()

    xpath_elem = f"//select[@id='exam_id']"
    exam_button = WebDriverWait(driver,25).until(EC.presence_of_element_located((By.XPATH, xpath_elem))).click()
    # exam_name = f"B.Sc. in Computer Science and Engineering 4th year 1st Semester Examination of 2022"
    # select_exam = driver.find_element(By.XPATH , '//select[@id="exam_id"]/option[contains(text(),exam_name)]')
    wait_function_until_stale()
    select_exams = driver.find_elements(By.XPATH , '//select[@id="exam_id"]/option')
    select_exams[1].click()

    css_elem = f"button[type='submit']"
    submit_but = WebDriverWait(driver,25).until(EC.presence_of_element_located((By.CSS_SELECTOR, css_elem)))
    actions.move_to_element(submit_but).click().perform()
    wait_function_until_stale()

    gpa_extractor()
    

reg_numbers = [i for i in range(681,740)]
radd_reg = [1771,1796,1758,1760,1811]
reg_numbers += radd_reg
for index,reg_no in enumerate(reg_numbers):
    print(reg_numbers[index])
    driver.switch_to.new_window('tab')
    driver.get(main_page)
    try:
        result_extractor(reg_no)
        final_reg.append(reg_no)
    except:
        print("result caanot be extracted trying next")
        pass
    driver.close()
    driver.switch_to.window(driver.window_handles[0])


# df=pd.DataFrame(news_dictionary)
# df1=df

# #sorting the dataframe by published time
# df1.sort_values(by='source_localtime' , ascending=False, inplace=True)
# df1.reset_index(drop=True, inplace=True)



# #checking if file alreaddy exists
# csv_filename = "Cuba_5_De_Septiembre.csv"
# if os.path.exists(csv_filename):
#     existing_df = pd.read_csv(csv_filename)
#     df1 = pd.concat([existing_df, df1], ignore_index=True)
#     df1 = df1.drop_duplicates(subset=["title"], keep="first")  
# else:
#     df1 = df1.drop_duplicates(subset=["title"], keep="first")

# #storing value in csv file
# df1.to_csv(csv_filename, index=False)
# df1
```
